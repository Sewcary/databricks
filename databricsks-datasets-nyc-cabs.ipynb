```python
# Lire les premières 100 lignes
file_path = "dbfs:/databricks-datasets/nyctaxi/tripdata/yellow/yellow_tripdata_2009-01.csv.gz"

df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(file_path) \
    .limit(10000)  # Charger seulement les 100 premières lignes

# Afficher un aperçu
df.show(10)
```

```python
# display(df)
from pyspark.sql.functions import col

df = df.filter("Trip_Distance > 1 ")
df=df.orderBy(col("trip_pickup_datetime").asc())
# display (df)

# df.printSchema()

df_new = df.select("vendor_name","trip_pickup_datetime","trip_dropoff_datetime","passenger_count","trip_distance","start_lon","start_lat","end_lon","end_lat","payment_type","fare_amt","total_amt")

df_new = df_new.withColumnRenamed("trip_pickup_datetime","trip_dropoff_datetime")
display (df_new)




```
